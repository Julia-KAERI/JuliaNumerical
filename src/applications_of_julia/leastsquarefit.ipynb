```julia
using LsqFit, Plots
```

```julia
τ = 3.0
x = 0.0:0.1:10.0
y = 10.0*exp.(-x ./ τ) .+ rand(length(x)).-0.5;
model(x, p) = p[1]*exp.(-x/p[2]) .+ p[3]
```

```julia
fit = curve_fit(model, collect(x), y, [5.0, 3.0, 1.0])
scatter(x, y, label = "Data", frame = :box)
plot!(x, model(x, fit.param), lw=2, color=:red, label = "Fit")
```
