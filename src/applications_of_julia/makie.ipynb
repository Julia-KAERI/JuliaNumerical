```julia

t=0.0:0.5:25.0
data = 4.0 .* exp.(-0.4 .* t) .+ (rand(length(t)) .- 0.5)
fig = Figure()
ax = Axis(fig[1, 1],
    title = "Exponential Decay and Fit",
    xlabel = "Time [Arb. Unit]",
    ylabel = "Value",
)
scatter!(
    ax,
    t,
    data,
    color = :tomato,
    label = "Measurements"
)
lines!(
    ax,
    t,
    4.0 .* exp.(-0.4 .* t),
    color = :blue,
    linestyle = :dash,
    label = L"f(x) = \exp(-0.4 t)",
)
axislegend(position = :rt)
fig
```

```julia
t=0.0:0.5:25.0
data = 4.0 .* exp.(-0.4 .* t) .+ (rand(length(t)) .- 0.5)
fig = Figure()
ax1 = Axis(fig[1, :],
    title = "Exponential Decay and Fit",
    xlabel = "Time [Arb. Unit]",
    ylabel = "Value",
)
scatter!(
    ax1,
    t,
    data,
    color = :tomato,
    label = "Measurements"
)
lines!(
    ax1,
    t,
    4.0 .* exp.(-0.4 .* t),
    color = :blue,
    linestyle = :dash,
    label = L"f(x) = \exp(-0.4 t)",
)
axislegend( position = :rt)

ax2 = Axis(fig[2, 1],
    title = "Image Plot",
)
image!(ax2, rand(100, 100))

ax3 = Axis(fig[2, 2],
    title = "Image Plot",
)
heatmap!(ax3, rand(100, 100))

fig
```
