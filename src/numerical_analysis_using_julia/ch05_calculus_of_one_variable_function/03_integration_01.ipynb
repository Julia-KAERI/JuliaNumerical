```julia
function integrate_trapzoidal(f::Function, a::Number, b::Number, n::Integer)
    a, b = minmax(a, b)
    h = (b-a)/n
    x = range(a, b, length = n+1)
    ff = f.(x)
    result = 0.5*(ff[1]+ff[end])
    result += sum(ff[2:end-1])
    return result*h
end

function integrate_simpson_1_3(f::Function, a::Number, b::Number, n::Integer)
    @assert n %2 ==0
    a, b = minmax(a, b)
    h = (b-a)/n
    x = range(a, b, length = n+1)
    ff = f.(x)
    result = (ff[1]+ff[end])
    result += 4*sum(ff[2:2:end-1])
    result += 2*sum(ff[3:2:end-2])
    return result * h/3
end

function integrate_simpson_3_8(f::Function, a::Number, b::Number, n::Integer)
    @assert n %3 ==0
    a, b = minmax(a, b)
    h = (b-a)/n
    x = range(a, b, length = n+1)
    println("h=$h, dx=$(x[10]-x[9])")
    ff = f.(x)
    result = (ff[1]+ff[end])
    result += 3*sum(ff[2:3:end-2])
    result += 3*sum(ff[3:3:end-1])
    result += 2*sum(ff[4:3:end-3])
    return result * h * 3 / 8
end
```

```julia
gg=x->sqrt(cos(x)+x)
nn = 40
rr = zeros(Float64, (nn, 3))
for i in 2:(1+nn)
    N=i*6
    rr[i-1, :] = [integrate_trapzoidal(gg, 0, 5, N) integrate_simpson_1_3(gg, 0, 5, N) integrate_simpson_3_8(gg, 0, 5, N)]
end
```

```julia
plot(6*(2:(1+nn)), abs.(rr[:,1] .- rr[end, 3]), lw=2, label = L"\textrm{Trapzoidal}", xlabel = "Number of subinterval", ylabel="abs(Err)")
plot!(6*(2:(1+nn)), abs.(rr[:,2].- rr[end, 3]), lw=2, label = L"\textrm{Simplson}_{1/3}")
plot!(6*(2:(1+nn)), abs.(rr[:,3].- rr[end, 3]), lw=2, label = L"\textrm{Simplson}_{3/8}")
```
