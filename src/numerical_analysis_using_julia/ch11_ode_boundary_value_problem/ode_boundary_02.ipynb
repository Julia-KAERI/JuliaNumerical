```julia
t, φ = linear_shooting(r->-2/r, r->0, r-> 0, [2, 10], [10, 0])
```

```julia
fig, ax= plot(t, φ, label = "Linear shooting")
t1 = 2.0:0.1:10
u = @. 10*2/t1 *(10-t1)/(8)
lines!(ax, t1, u, color = :red)
ax.xlabel = L"r"
ax.ylabel = L"\varphi"
fig
```
