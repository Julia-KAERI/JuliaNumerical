```julia
function ode_euler(f::Function, x0::Real, y0::Real, Npoints::Integer, h = 1.0e-6)
    x0, y0, h = promote(x0, y0, h)
    xp = x0 .+ collect(0:1:(Npoints-1)) * h
    yp = zero(xp)
    yp[1] = y0
    for i in 2:Npoints
        yp[i] = yp[i-1]+ f(xp[i-1]) * h
    end
    return xp, yp
end
```

```julia
xp, yp = ode_euler(exp, 0.0, 1.0, 50, 1.0e-1)
plot(xp, yp)
```
