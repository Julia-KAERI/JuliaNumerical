```julia
t=-40:0.01:40
pp(t) = (abs(t)<0.0001) ? 0.0 : (sin(t/4))^2/(t/4)
y = pp.(t)
f1 = Figure()
ax1 = Axis(f1[1, 1])
scatter!(ax1, t, abs.(y))
f1
```

```julia
ϕ(t) = (t< 0) ? 0 : (t> 1) ? 0 : 1
Haar(t) = ( 0 ≤ t < 0.5) ? 1 : ( 0.5 ≤ t < 1) ? -1 : 0
```

```julia
s = -2:0.01:2
f = Figure()
ax = Axis(f[1, 1])
lines!(ax, s, ϕ.(s), label = L"\phi(t)")
lines!(ax, s, Haar.(s), label = "Haar")
axislegend()
f
```

```julia
s=-5:0.01:5
phis = @. exp(-s^2)
haars = Haar.(s)
t=conv(phis, haars)
f3 = Figure()
ax3= Axis(f3[1, 1])
ax3.xlabel = L"\omega"
ax3.title = L"\mathfrak{F}[\psi_H \ast \phi](t)"
lines!(ax3, s, t[500:end-501])
f3
# save("wavelet_2.png", f3)
```
